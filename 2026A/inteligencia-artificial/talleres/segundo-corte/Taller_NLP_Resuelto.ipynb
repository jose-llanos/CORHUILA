## **Ejercicio 1: Limpieza y Preprocesamiento Básico**

**Objetivo:** Crear una función que limpie texto removiendo URLs, emails, caracteres especiales y normalizando espacios.

**Instrucciones:**
1. Implementa la función `limpiar_texto()` que:
   - Convierte a minúsculas
   - Remueve URLs (http, https, www)
   - Remueve emails
   - Remueve caracteres especiales manteniendo solo letras y espacios
   - Normaliza espacios en blanco

2. Prueba con los siguientes textos:

```python
import re

def limpiar_texto(texto):
    """
    Limpia un texto removiendo elementos no deseados.
    
    Args:
        texto: String con el texto a limpiar
    
    Returns:
        String con el texto limpio
    """
    # Convertir a minúsculas
    texto = texto.lower()
    
    # Remover URLs
    texto = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', texto)
    texto = re.sub(r'www\.[a-zA-Z0-9.-]+\.[a-zA-Z]{2,}', '', texto)
    
    # Remover emails
    texto = re.sub(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', '', texto)
    
    # Remover caracteres especiales, mantener solo letras y espacios
    texto = re.sub(r'[^a-záéíóúüñ\s]', '', texto)
    
    # Normalizar espacios
    texto = re.sub(r'\s+', ' ', texto).strip()
    
    return texto

# Casos de prueba
textos_prueba = [
    "¡Hola!!! Visita https://ejemplo.com o envía tu CV a recurso@empresa.com. Hay   espacios extra!!!",
    "Python3.9 es excelente!!! Descarga desde www.python.org #programming @user",
    "El NLP es fascinante... ¿No lo crees? (sí/no) [responder]"
]

print("Resultados de limpieza:")
print("=" * 60)
for i, texto in enumerate(textos_prueba, 1):
    print(f"\nTexto {i}:")
    print(f"Original:  {texto}")
    print(f"Limpio:    {limpiar_texto(texto)}")
```

## **Ejercicio 2: Tokenización**

**Objetivo:** Dividir textos en tokens (palabras u oraciones) utilizando NLTK.

**Instrucciones:**
1. Descarga los recursos necesarios de NLTK
2. Implementa tokenización de palabras y oraciones
3. Compara los resultados para diferentes textos

```python
import nltk

# Descargar recursos necesarios
nltk.download('punkt_tab')
nltk.download('punkt')

from nltk.tokenize import word_tokenize, sent_tokenize

def analizar_tokenizacion(texto, idioma='spanish'):
    """
    Realiza tokenización de un texto en palabras y oraciones.
    
    Args:
        texto: String con el texto a tokenizar
        idioma: Idioma del texto
    
    Returns:
        Diccionario con tokens de palabras y oraciones
    """
    tokens_palabra = word_tokenize(texto, language=idioma)
    tokens_oracion = sent_tokenize(texto, language=idioma)
    
    resultado = {
        'tokens_palabra': tokens_palabra,
        'tokens_oracion': tokens_oracion,
        'num_palabras': len(tokens_palabra),
        'num_oraciones': len(tokens_oracion)
    }
    return resultado

# Casos de prueba
texto1 = "Hola. ¿Cómo estás? Estoy muy bien, gracias por preguntar."
texto2 = "El procesamiento del lenguaje natural es increíble! ¿No crees? Sí, totalmente."

for i, texto in enumerate([texto1, texto2], 1):
    print(f"\nTexto {i}: {texto}")
    resultado = analizar_tokenizacion(texto)
    print(f"Tokens de palabras ({resultado['num_palabras']}): {resultado['tokens_palabra']}")
    print(f"Tokens de oraciones ({resultado['num_oraciones']}): {resultado['tokens_oracion']}")
```

## **Ejercicio 3: Normalización de Texto**

**Objetivo:** Normalizar texto removiendo acentos y caracteres diacríticos.

**Instrucciones:**
1. Implementa función que remueva acentos usando `unicodedata`
2. Convierte textos con acentos a su forma normalizada
3. Compara el antes y después

```python
import unicodedata

def normalizar_texto(texto):
    """
    Normaliza un texto removiendo acentos y caracteres diacríticos.
    
    Args:
        texto: String con el texto a normalizar
    
    Returns:
        String normalizado
    """
    # Convertir a minúsculas
    texto = texto.lower()
    # Normalizar acentos
    texto = unicodedata.normalize('NFD', texto)
    texto = ''.join(char for char in texto if unicodedata.category(char) != 'Mn')
    return texto

# Casos de prueba
textos_acentuados = [
    "¡Hola! ¿Cómo Estás?",
    "Programación y árboles de decisión",
    "La física y química de la Tierra",
    "Café con leche es delicioso"
]

print("Normalización de textos acentuados:")
print("=" * 60)
for texto in textos_acentuados:
    print(f"Original:     {texto}")
    print(f"Normalizado:  {normalizar_texto(texto)}")
    print()
```

## **Ejercicio 4: Remoción de Stopwords**

**Objetivo:** Remover palabras comunes que no aportan significado.

**Comentarios:**
- Usamos la lista `stopwords` de NLTK para español.
- Se tokeniza y se filtra cada palabra contra el conjunto de stopwords.
- Se conserva el resto de tokens relevantes.


```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords', quiet=True)

# Función para remover stopwords incluyendo limpieza básica

def remover_stopwords(texto, idioma='spanish'):
    """Remueve stopwords de un texto en español."""
    texto_limpio = limpiar_texto(texto)  # reutilizamos limpieza básica del Ejercicio 1
    tokens = word_tokenize(texto_limpio, language=idioma)
    stop_words = set(stopwords.words(idioma))

    tokens_filtrados = [token for token in tokens if token not in stop_words]
    return tokens_filtrados

# Ejemplos de uso
textos = [
    "El procesamiento del lenguaje natural es fascinante y útil",
    "La inteligencia artificial está transformando el mundo",
    "Los algoritmos de aprendizaje automático son muy poderosos"
]

print("Remoción de Stopwords:")
print("=" * 60)
for texto in textos:
    resultado = remover_stopwords(texto)
    print(f"Original:      {texto}")
    print(f"Sin stopwords: {' '.join(resultado)}")
    print(f"Palabras removidas: {len(texto.split()) - len(resultado)}")
    print()
```

## **Ejercicio 5: Stemming vs Lemmatización**

**Objetivo:** Comparar dos técnicas de reducción de palabras.

**Comentarios:**
- SnowballStemmer funciona bien en español y reduce cada palabra a su raíz (no siempre léxicamente exacta).
- WordNetLemmatizer proviene de WordNet (inglés) y devuelve formas canónicas, especialmente útil con POS.


```python
from nltk.stem.snowball import SnowballStemmer
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)

# Comparación directa con impresión

def comparar_stemming_lemmatizacion(palabras_es, palabras_en):
    stemmer = SnowballStemmer('spanish')
    lemmatizer = WordNetLemmatizer()

    print("Comparación: Stemming (Español)")
    print("=" * 60)
    for palabra in palabras_es:
        stem = stemmer.stem(palabra)
        print(f"{palabra:<15} -> {stem}")

    print("\nComparación: Lemmatización (Inglés)")
    print("=" * 60)
    for palabra in palabras_en:
        lema = lemmatizer.lemmatize(palabra, pos='v')
        print(f"{palabra:<15} -> {lema}")

    print("\nAnálisis:")
    print("- Stemming es rápido pero puede producir tokens que no son palabras reales.")
    print("- Lemmatización es más precisa semanticamente y depende de POS y diccionario.")

palabras_es = [
    'correr', 'corriendo', 'corrió', 'corrida',
    'programar', 'programación', 'programador',
    'análisis', 'analizar', 'analizando'
]

palabras_en = [
    'running', 'runs', 'ran',
    'programming', 'programs', 'programmed',
    'analyzing', 'analyzed', 'analyzes'
]

comparar_stemming_lemmatizacion(palabras_es, palabras_en)
```

## **Ejercicio 6: Análisis de Frecuencia de Palabras**

**Objetivo:** Identificar las palabras más frecuentes en un texto.

**Comentarios:**
- Limpiamos el texto, tokenizamos y filtramos stopwords.
- Contamos frecuencias con `Counter` y mostramos porcentaje del corpus.


```python
from collections import Counter


def obtener_palabras_clave(texto, n=10, idioma='spanish'):
    """Extrae las N palabras más frecuentes de un texto sin stopwords."""
    texto = normalizar_texto(texto)
    texto = re.sub(r'[^a-zñ\s]', ' ', texto)
    tokens = texto.split()

    stop_words = set(stopwords.words(idioma))
    tokens_filtrados = [t for t in tokens if t not in stop_words and len(t) > 1]

    contador = Counter(tokens_filtrados)
    return contador.most_common(n), len(tokens_filtrados)

# Texto de ejemplo
texto_analisis = """
La Inteligencia Artificial es revolucionaria. La IA transforma industrias.
El aprendizaje automático es clave. Los modelos de IA mejoran constantemente.
Los algoritmos de machine learning procesan datos masivos.
La IA encontrada en aplicaciones modernas es muy sofisticada.
Programación y IA son futuro del software.
"""

print("Análisis de Palabras Clave")
print("=" * 60)
resultado, total_tokens = obtener_palabras_clave(texto_analisis, n=10)

print(f"Total de palabras (descontando stopwords): {total_tokens}\n")
print(f"{'Palabra':<15} {'Frecuencia':<12} {'Porcentaje':<10}")
print("-" * 45)
for palabra, freq in resultado:
    porcentaje = (freq / total_tokens * 100) if total_tokens else 0
    print(f"{palabra:<15} {freq:<12} {porcentaje:.2f}%")
```

## **Ejercicio 7: Extracción de Entidades Nombradas (NER)**

**Objetivo:** Identificar entidades como personas, lugares y organizaciones.

**Comentarios:**
- Usamos spaCy con el modelo `es_core_news_sm`.
- `extraer_entidades` devuelve lista de entidades con texto y etiqueta.
- Agrupamos por tipo para análisis.

```python
import spacy

# Cargar modelo de spaCy una vez (punto 7)
nlp_es = spacy.load('es_core_news_sm')


def extraer_entidades(texto):
    """Extrae entidades nombradas con spaCy."""
    doc = nlp_es(texto)
    entidades = [{'texto': ent.text, 'tipo': ent.label_} for ent in doc.ents]
    return entidades

# Textos de ejemplo para NER
textos_ner = [
    """
    Juan García, director ejecutivo de Google, anunció hoy en Bogotá
    la apertura de una oficina regional. La inversión será de 50 millones
    de dólares en los próximos 3 años.
    """,
    """
    María López trabaja en Microsoft desde el 2020. Ella es ingeniera
    de software y reside en Seattle, Washington. Su proyecto actual
    es sobre inteligencia artificial.
    """
]

print("Extracción de Entidades (NER)")
print("=" * 70)

for i, texto in enumerate(textos_ner, 1):
    print(f"\nTexto {i}:")
    print(texto.strip())
    entidades = extraer_entidades(texto)

    if entidades:
        for ent in entidades:
            print(f"- Entidad: '{ent['texto']}', Tipo: {ent['tipo']}")
    else:
        print("No se encontraron entidades.")
```

## **Ejercicio 8: Similaridad entre Textos**

**Objetivo:** Calcular qué tan similares son dos textos usando similitud coseno.

**Comentarios:**
- TF-IDF transforma los textos en vectores numéricos.
- La similitud coseno muestra valores entre 0 (diferente) y 1 (idéntico).


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np


def calcular_similitud_textos(textos):
    """Calcula la matriz de similitud coseno entre varios textos."""
    vectorizer = TfidfVectorizer(lowercase=True, stop_words='spanish')
    tfidf = vectorizer.fit_transform(textos)
    sim = cosine_similarity(tfidf)
    return sim

# Textos de prueba de similitud
textos_similitud = [
    "El procesamiento del lenguaje natural es importante para la IA",
    "NLP y procesamiento de texto son relevantes en inteligencia artificial",
    "Los gatos son animales domésticos muy populares",
    "La comprensión del lenguaje natural mediante máquinas es fundamental",
    "Los perros y gatos son mascotas comunes en los hogares"
]

matriz = calcular_similitud_textos(textos_similitud)

print("Análisis de Similitud entre Textos")
print("=" * 80)

n = len(textos_similitud)
print(f"{'':<6}" + ''.join([f"T{i+1:<10}" for i in range(n)]))
for i in range(n):
    fila = ''.join([f"{matriz[i,j]:.2f}     " for j in range(n)])
    print(f"T{i+1:<3} {fila}")

print("\nInterpretación: Valores cercanos a 1 indican textos similares.")
```

## **Ejercicio 9: Pipeline Completo de Preprocesamiento**

**Objetivo:** Crear una clase que integre todas las técnicas anteriores.

**Comentarios:**
- Implementamos la clase `PreprocessadorTexto` con métodos privados y `procesar`.
- Paso a paso: limpieza, normalización, tokenización, stopwords, stemming.
- Permite reutilizar el conjunto de funciones de los ejercicios previos.


```python
class PreprocessadorTexto:
    def __init__(self, idioma='spanish'):
        self.idioma = idioma
        self.stemmer = SnowballStemmer(idioma)
        self.stopwords_set = set(stopwords.words(idioma))

    def _limpiar(self, texto):
        # Reinicia pipeline con limpieza completa
        return limpiar_texto(texto)

    def _normalizar(self, texto):
        return normalizar_texto(texto)

    def _tokenizar(self, texto):
        return word_tokenize(texto, language=self.idioma)

    def _remover_stopwords(self, tokens):
        return [tok for tok in tokens if tok not in self.stopwords_set]

    def _aplicar_stemming(self, tokens):
        return [self.stemmer.stem(tok) for tok in tokens]

    def procesar(self, texto, verbose=False):
        texto_limpio = self._limpiar(texto)
        texto_normalizado = self._normalizar(texto_limpio)
        tokens = self._tokenizar(texto_normalizado)
        tokens_no_stop = self._remover_stopwords(tokens)
        tokens_stem = self._aplicar_stemming(tokens_no_stop)

        if verbose:
            print("Texto original:\n", texto.strip())
            print("\nTexto limpio:", texto_limpio)
            print("\nTexto normalizado:", texto_normalizado)
            print("\nTokens:", tokens)
            print("\nDespués de stopwords:", tokens_no_stop)
            print("\nTokens con stemming:", tokens_stem)

        return tokens_stem

# Prueba del pipeline completo
preprocesador = PreprocessadorTexto()
for i, texto in enumerate(textos_complejos, 1):
    print(f"\n\n{'='*40}\nEjemplo {i}\n{'='*40}")
    resultado_pipeline = preprocesador.procesar(texto, verbose=True)
    print(f"Resultado final tokens ({len(resultado_pipeline)}):", resultado_pipeline)
```

## **Ejercicio 10: Proyecto Integrador - Análisis de Sentimientos Básico**

**Objetivo:** Integrar todas las técnicas para un proyecto simple de análisis de sentimientos.

**Comentarios:**
- Se define un analizador basado en similitud TF-IDF entre el texto y una lista de palabras clave de sentimientos.
- Clasifica sentimiento en positivo, negativo o neutral con confianza relativa.


```python
class AnalizadorSentimientos:
    def __init__(self):
        # Listas de palabras clave de sentimiento.
        self.palabras_positivas = "excelente maravilloso fantástico recomendado feliz bueno positivo".split()
        self.palabras_negativas = "malo terrible horrible decepcionante frustrante pésimo negativo".split()
        self.palabras_neutras = "normal regular promedio estable".split()

    def _score_por_clave(self, texto, claves):
        vectorizer = TfidfVectorizer(lowercase=True, stop_words='spanish')
        docs = [texto] + [' '.join(claves)]
        tfidf = vectorizer.fit_transform(docs)
        score = cosine_similarity(tfidf[0:1], tfidf[1:2])[0, 0]
        return score

    def analizar(self, texto):
        texto_limpio = normalizar_texto(limpiar_texto(texto))

        sim_pos = self._score_por_clave(texto_limpio, self.palabras_positivas)
        sim_neg = self._score_por_clave(texto_limpio, self.palabras_negativas)
        sim_neu = self._score_por_clave(texto_limpio, self.palabras_neutras)

        max_sim = max(sim_pos, sim_neg, sim_neu)
        if max_sim == sim_pos:
            sentimiento = 'positivo'
        elif max_sim == sim_neg:
            sentimiento = 'negativo'
        else:
            sentimiento = 'neutral'

        return {
            'texto': texto,
            'sentimiento': sentimiento,
            'confianza': float(max_sim),
            'similitud_positiva': float(sim_pos),
            'similitud_negativa': float(sim_neg),
            'similitud_neutral': float(sim_neu)
        }

# Análisis de opiniones
analizador = AnalizadorSentimientos()

print("Análisis de Sentimientos")
print("=" * 80)
print(f"{'Opinión':<55} {'Sentimiento':<10} {'Confianza':<10}")
print("-" * 80)
for texto in opiniones:
    resultado = analizador.analizar(texto)
    resumen = texto[:50] + ('...' if len(texto) > 50 else '')
    print(f"{resumen:<55} {resultado['sentimiento']:<10} {resultado['confianza']:.2f}")
```

## **Desafío Final**

**Objetivo:** Integrar todo el taller en un mini-proyecto.

**Comentarios:**
- Se procesa un conjunto de reseñas, se extraen entidades, se calcula sentimiento y se genera un reporte resumido.
- La función `reporte_final` usa los componentes creados de secciones previas.


```python
from collections import defaultdict

# Desafío final: procesamiento integral

def reporte_final(reseñas):
    pre = PreprocessadorTexto()
    analizador = AnalizadorSentimientos()
    entidades_contador = Counter()
    conteo_sentimientos = Counter()

    for r in reseñas:
        resultado = analizador.analizar(r)
        conteo_sentimientos[resultado['sentimiento']] += 1

        for ent in extraer_entidades(r):
            entidades_contador[ent['texto']] += 1

    sentimiento_promedio = (conteo_sentimientos['positivo'] - conteo_sentimientos['negativo']) / len(reseñas)

    print("Reporte Final")
    print("=" * 80)
    print(f"Total de reseñas: {len(reseñas)}")
    print("Conteo por sentimiento:", dict(conteo_sentimientos))
    print(f"Sentimiento promedio (positivo-negativo): {sentimiento_promedio:.2f}")
    print("Entidades más mencionadas:", entidades_contador.most_common(10))

    # Similaridad entre reseñas para agrupar
    vec = TfidfVectorizer(lowercase=True, stop_words='spanish')
    matriz = cosine_similarity(vec.fit_transform([limpiar_texto(r) for r in reseñas]))

    # Agrupado simple: reseñas con similitud > 0.5 se consideran juntas
    grupos = []
    ya_encontrados = set()

    for i in range(len(reseñas)):
        if i in ya_encontrados:
            continue
        grupo = [i]
        ya_encontrados.add(i)
        for j in range(i + 1, len(reseñas)):
            if matriz[i, j] > 0.50:
                grupo.append(j)
                ya_encontrados.add(j)
        grupos.append(grupo)

    print("Grupos de reseñas similares (índices):", grupos)


test_reseñas = [
    "Este producto es excelente y me encanta. Muy recomendado!",
    "Horrible, defectuoso y de mala calidad. Totalmente decepción.",
    "El producto llegó a tiempo. Normal, sin sorpresas.",
    "¡Increíble! Superó todas mis expectativas. Fantástico!",
    "No funciona bien, es frustrante y de mala calidad."
]

reporte_final(test_reseñas)
```
